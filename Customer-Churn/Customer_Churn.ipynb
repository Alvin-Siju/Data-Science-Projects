```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Monthly','Quarterly','Annual']])
```

```python
df['Contract Length']=oe.fit_transform(df[['Contract Length']])
```

```python
cat_cols
```

```python
df=pd.get_dummies(df,columns=['Gender','Subscription Type'],drop_first=True)
```

```python
X=df.drop(['Churn','CustomerID'],axis=1)
y=df['Churn']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
sss=StratifiedShuffleSplit(n_splits=1,test_size=0.3,random_state=42)
for train,test in sss.split(X,y):
    X_train,y_train=X.iloc[train],y.iloc[train] 
    X_test,y_test=X.iloc[test],y.iloc[test]
```

```python
estimator=Pipeline([('scaler',StandardScaler()),
                    ('dectree',DecisionTreeClassifier(criterion='gini'))
                    ])
params={"dectree__max_depth":[3,4,5,6,7,8,9,10],
        "dectree__max_features":[3,6,9,13]}
```

```python
grid=GridSearchCV(estimator,params,cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_,grid.best_score_,grid.best_estimator_)
```

```python
model=grid.best_estimator_
```

```python
pred=model.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred),"\n")
print(classification_report(y_test,pred))
```
